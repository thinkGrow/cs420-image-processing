The answer to the question -> The actual function

```python
def intensityMultiplier(scaling_factor,im_np):
    im_np = im_np * scaling_factor
    return im_np
```

Experimenting with different scaling factors for our function and plotting them

```python
im2_np = intensityMultiplier(2,im_np)
im3_np = intensityMultiplier(3,im_np)
im4_np = intensityMultiplier(4,im_np)
```

```python
# code for displaying multiple images in one figure
    
# create figure
fig = plt.figure(figsize=(10, 7))
  
# setting values to rows and column variables
rows = 2
columns = 2
   
# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
  
# showing image
plt.imshow(im_np, "gray")
plt.axis('off')
plt.title("Original")
  
# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)
  
# showing image
plt.imshow(im2_np, "gray")
plt.axis('off')
plt.title("Scaled Image by 2 factor")

# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)
  
# showing image
plt.imshow(im3_np, "gray")
plt.axis('off')
plt.title("Scaled Image by 3 factor")

# Adds a subplot at the 4th position
fig.add_subplot(rows, columns, 4)
  
# showing image
plt.imshow(im4_np, "gray")
plt.axis('off')
plt.title("Scaled Image by 4 factor");
```
